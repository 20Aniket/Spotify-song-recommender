```python
from dotenv import load_dotenv
import spotipy
from spotipy.oauth2 import SpotifyOAuth
import os
import pandas as pd
```

```python
from IPython.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
```

```python
load_dotenv()
Client_ID= os.getenv("CLIENT_ID")
Client_Secret= os.getenv("CLIENT_SECRET")
```

```python
#setup for authentication
sp= spotipy.Spotify(auth_manager=SpotifyOAuth(client_id=Client_ID, 
                                          client_secret=Client_Secret, 
                                          redirect_uri="http://localhost:8000", 
                                          scope= "user-read-recently-played playlist-read-private playlist-modify-public playlist-modify-private"))
```

```python
# Get users 50 most played tracks
results = sp.current_user_recently_played(limit=50)

data = []
for item in results['items']:
    track = item['track']
    features = sp.audio_features(track['id'])[0]
    acousticness = features['acousticness']
    danceability = features['danceability']
    energy = features['energy']
    instrumentalness = features['instrumentalness']
    liveness = features['liveness']
    loudness = features['loudness']
    tempo = features['tempo']

    # Get the genre information for the track
    artist_id = track['artists'][0]['id']
    artist = sp.artist(artist_id)
    genres = artist['genres']

    data.append([acousticness, danceability, energy, instrumentalness, liveness, loudness, tempo, genres])

df = pd.DataFrame(data, columns=['Acousticness', 'Danceability', 'Energy', 'Instrumentalness', 'Liveness', 'Loudness', 'Tempo', 'Genres'])
df.to_csv('spotify_data.csv', index=False)

# Find the most common genre
genre_counts = df['Genres'].explode().value_counts()
most_common_genre = genre_counts.idxmax()

print('Most common genre:', most_common_genre)
```

```python
import pandas as pd
spotify_df = pd.read_csv('spotify_data.csv')
min_Instrumentalness= spotify_df['Instrumentalness'].min()
max_Instrumentalness= spotify_df['Instrumentalness'].max()
min_Acousticness= spotify_df['Acousticness'].min()
max_Acousticness= spotify_df['Acousticness'].max()
min_Danceability= spotify_df['Danceability'].min()
max_Danceability= spotify_df['Danceability'].max()
min_Liveness= spotify_df['Liveness'].min()
max_Liveness= spotify_df['Liveness'].max()
min_Loudness= spotify_df['Loudness'].min()
max_Loudness= spotify_df['Loudness'].max()
min_Energy= spotify_df['Energy'].min()
max_Energy= spotify_df['Energy'].max()
min_Tempo= spotify_df['Tempo'].min()
max_Tempo= spotify_df['Tempo'].max()

# Create an empty DataFrame to store the tracks that meet the criteria
filtered_tracks = pd.DataFrame(columns=['ID', 'Name', 'Artist', 'Acousticness', 'Danceability', 'Energy', 'Instrumentalness', 'Liveness', 'Loudness', 'Tempo'])
for playlist in playlists['playlists']['items']:
    # Get the tracks in the playlist
    tracks = sp.playlist_tracks(playlist['id'])
    for item in tracks['items']:
        track = item['track']
        track_id = track['id']
        name = track['name']
        artist = track['artists'][0]['name']
        features = sp.audio_features(track_id)[0]
        acousticness = features['acousticness']
        danceability = features['danceability']
        energy = features['energy']
        instrumentalness = features['instrumentalness']
        liveness = features['liveness']
        loudness = features['loudness']
        tempo = features['tempo']
        if (acousticness >= min_Acousticness and acousticness <= max_Acousticness and
            danceability >= min_Danceability and danceability <= max_Danceability and
            energy >= min_Energy and energy <= max_Energy and
            instrumentalness >= min_Instrumentalness and instrumentalness <= max_Instrumentalness and
            liveness >= min_Liveness and liveness <= max_Liveness and
            loudness >= min_Loudness and loudness <= max_Loudness and
            tempo >= min_Tempo and tempo <= max_Tempo):
            # Add the track to the DataFrame
            new_row = pd.DataFrame({'ID': [track_id], 'Name': [name], 'Artist': [artist], 'Acousticness': [acousticness], 'Danceability': [danceability], 'Energy': [energy], 'Instrumentalness': [instrumentalness], 'Liveness': [liveness], 'Loudness': [loudness], 'Tempo': [tempo]})
            filtered_tracks = pd.concat([filtered_tracks, new_row], ignore_index=True)
            filtered_tracks.to_csv('recommendations.csv', index=False)
```
